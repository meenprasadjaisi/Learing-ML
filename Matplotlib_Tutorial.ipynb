## Matplotlib Tutorial

. Useful for making Plots




```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
# Plot sine wave
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine Wave')
plt.show()
```

```python
# Plot Cosine wave
plt.plot(x,z)
plt.xlabel('x')
plt.ylabel('z')
plt.title('Cosine Wave')
plt.show()
```

```python
# Parabola
x = np.linspace(-10,10,30)
y = x**2
plt.plot(x,y, 'r*')
plt.title('Parabola')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

Ploting Two Graphs

```python
x = np.linspace(0,15,85)
plt.plot(x, np.sin(x), 'r--')
plt.plot(x, np.cos(x), 'g')
plt.show()
```

Bar Graph Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.bar(languages, people)
plt.xlabel('Languages')
plt.ylabel('People')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
